##Laplacian Operator

```python
# Importing modules
import cv2
import matplotlib.pyplot as plt

# Reading the image
img = cv2.imread('dancing-spider.jpeg')
```

```python
# Applying gaussian filter
blur_img = cv2.GaussianBlur(img, (3, 3), 0)
```

```python
# Converting image to grayscale
gray_img = cv2.cvtColor(blur_img, cv2.COLOR_BGR2GRAY)
```

```python
# Laplacian Operator
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F, ksize=1)
abs_laplacian = cv2.convertScaleAbs(laplacian)
```

```python
plt.figure()
plt.title('Spider - Laplacian')
plt.imsave('lap.png', laplacian, cmap='gray', format='png')
plt.imshow(abs_laplacian, cmap='gray')
plt.show()
```

##Canny Operator

```python
# Applying Canny
edges = cv2.Canny(blur_img, 100, 200, 3, L2gradient=True)
```

```python
plt.figure()
plt.title('Spider - Canny')
plt.imsave('dancing-spider-canny.png', edges, cmap='gray', format='png')
plt.imshow(edges, cmap='gray')
plt.show()
```
